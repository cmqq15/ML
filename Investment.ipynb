```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv(r"D:\Dropbox\DS & AI\Full stack DS and AI course\20 Jun\Investment.csv")

X = dataset.iloc[:, :-1]
y = dataset.iloc[:, 4]

X = pd.get_dummies(X,dtype=int) # convert dummies into numbers

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)

y_pred = regressor.predict(X_test)

bias = regressor.score(X_train, y_train)
print(bias) 

# bias is 95

variance = regressor.score(X_test,y_test)
print(variance)

# variance is 93
# in this dataset, constant is missing
# Let's take it as 1
# Let us append 1 to every x.

X = np.append(arr = np.ones((50,1)).astype(int), values = X, axis =1)

```

```python
# New concept now, statsmodels.api (statsmodels.org)
# see at statsmodels.org/stable/api.html

import statsmodels.api as sm
X_opt = X[:,[0,1,2,3,4,5]]
# api means application programming interface
# sm means statsmodels
#OrdinaryLeastSquares
regressor_OLS = sm.OLS(endog=y, exog=X_opt).fit()
regressor_OLS.summary()
```
